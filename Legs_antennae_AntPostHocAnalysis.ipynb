<h1><font color="blue">Cluster (GMM)</font></h1>

```python
from sklearn.mixture import GaussianMixture

matplotlib.rc('font', size=MEDIUM_SIZE)
matplotlib.rc('axes', titlesize=MEDIUM_SIZE)
```

```python
# *** DEFINE ***
max_cluster = 20

# Compute BIC
cluster_list = range(1, max_cluster)
bic_list = []
for n_components in tqdm(cluster_list):
    gmm = GaussianMixture(n_components)
    gmm.fit(tot_embed_valid)
    bic_list.append(gmm.bic(tot_embed_valid))
min_idx = np.argmin(bic_list)

# Plot
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(cluster_list, bic_list, marker="o", c='k', markersize=5)
# ax.scatter(min_idx+1, bic_list[min_idx], s=120, edgecolors='r', facecolors='none')
ax.set_xticks(range(1, max_cluster))
ax.ticklabel_format(style="sci", scilimits=(0,5))
ax.set(xlabel='Number of Clusters', ylabel='BIC', title="Bayesian Information Criterion")
plt.show()
```

```python
# Gaussian Mixture Model
gmm = GaussianMixture(n_components=15, random_state=5, max_iter=10000)
gmm_label = gmm.fit_predict(tot_embed_valid)

# Compute Stats
num_clusters = np.max(gmm_label) + 1
prob = np.max(gmm.predict_proba(tot_embed_valid), axis=1)

# Choose Color Palette
color_palette = sns.color_palette('rainbow', num_clusters)
cluster_colors = [color_palette[x] for x in gmm_label]
cluster_member_colors = np.array([sns.desaturate(x, p) for x, p in zip(cluster_colors, prob)])

# Create Figures
fig, ax = plt.subplots(figsize=(8, 8))
for i in range(num_clusters):
    idx = (gmm_label==i)
    ax.scatter(tot_embed_valid[idx,0], tot_embed_valid[idx,1], 
                c=cluster_member_colors[idx], 
                alpha=1, s=1, label=i)
    ax.annotate(i, gmm.means_[i], fontsize=14, fontweight='bold')
    
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', markerscale=3)
ax.set(xlabel="Component 1", ylabel="Component 2", title="Behavioral Manifold")
plt.show()
```

```python
import numpy as np
import matplotlib, random
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import seaborn as sns
from tqdm.notebook import tqdm

# frame info
# analyze_cluster = 11
# 
class Video:
    def __init__(self, bodypoints, angles, freq, power, embed, cluster, cluster_mean, fps=10, dpi=200):
        self.bodypoints = bodypoints
        self.embed = embed
        self.angles = angles
        self.freq = freq
        self.power = power
        self.cluster = cluster.astype(int)
        self.cluster_mean = cluster_mean
        self.fps = fps
        self.dpi = dpi
        
        self.good_fr = np.unique(np.where(~np.isnan(embed))[0])
        
        SMALL_SIZE = 4
        matplotlib.rc('font', size=SMALL_SIZE)
        matplotlib.rc('axes', titlesize=SMALL_SIZE)
        
        self.setup()

    def setup(self):
        # video info
        self.fig = plt.figure(constrained_layout=True)
        gs = self.fig.add_gridspec(16, 24)
        self.ax1 = self.fig.add_subplot(gs[0:8, 0:8])
        self.ax2 = self.fig.add_subplot(gs[0:4, 8:])
        self.ax3 = self.fig.add_subplot(gs[4:8, 8:])
        self.ax4 = self.fig.add_subplot(gs[8:, 0:8])
        self.ax5 = self.fig.add_subplot(gs[8:, 8:16])
        self.ax6 = self.fig.add_subplot(gs[8:, 16:24])

        # color palette
        num_clusters = np.max(self.cluster)+1
        color_palette = sns.color_palette('rainbow', int(num_clusters))
        self.cluster_colors = np.array([color_palette[x] for x in self.cluster])
        
        self.bp_palette = sns.color_palette('tab10', 10)
        
        # density cluster plot
        sns.kdeplot(self.embed[self.good_fr ,0], self.embed[self.good_fr ,1], shade=True, shade_lowest=True, 
                    gridsize=100, levels=40, cmap='jet',cbar=False, ax=self.ax6)
        for clust_i in range(num_clusters):
            self.ax6.annotate(clust_i, self.cluster_mean[clust_i], fontsize=4, fontweight='bold', color='k')
        self.ax6.set(xlabel='Component 1', ylabel='Component 2')


    def create_video_legs(self, start, stop, filepath):
        num_clusters = np.max(self.cluster)+1

        # scatter cluster plot
        samp_frac = 0.5
        num_fr, num_angles = self.angles.shape
        num_angles
        idx = random.sample(self.good_fr.tolist(), int(samp_frac*num_fr))

        FFMpegWriter = animation.writers['ffmpeg']
        writer = FFMpegWriter(fps=self.fps)
        with writer.saving(self.fig, filepath, dpi=self.dpi):
            for fr_i, fr in enumerate(tqdm(np.arange(start, stop), desc="Frame Loop")):
                # ant plot
                self.ax1.clear()
                self.ax1.set_xlim([-6,6]); self.ax1.set_ylim([-5,8]);
                for shadow_i in range(-10,1):
                    if shadow_i == 0:
                        alpha=0.8
                    else:
                        alpha = 0.1
                    self.ax1.plot(self.bodypoints[fr+shadow_i,0:4,0], self.bodypoints[fr+shadow_i,0:4,1], 
                             c=self.bp_palette[0], alpha=alpha, 
                             marker="o", markersize=1)
                    
                    self.ax1.plot(self.bodypoints[fr+shadow_i,4:8,0], self.bodypoints[fr+shadow_i,4:8,1], 
                             c=self.bp_palette[1], alpha=alpha, 
                             marker="o", markersize=1)
                    self.ax1.plot(self.bodypoints[fr+shadow_i,8:11,0], self.bodypoints[fr+shadow_i,8:11,1], 
                             c=self.bp_palette[2], alpha=alpha, 
                             marker="o", markersize=1)
                    self.ax1.plot(self.bodypoints[fr+shadow_i,11:14,0], self.bodypoints[fr+shadow_i,11:14,1], 
                             c=self.bp_palette[3], alpha=alpha, 
                             marker="o", markersize=1)
                    self.ax1.plot(self.bodypoints[fr+shadow_i,14:17,0], self.bodypoints[fr+shadow_i,14:17,1], 
                             c=self.bp_palette[4], alpha=alpha, 
                             marker="o", markersize=1)
                    
                    self.ax1.plot(self.bodypoints[fr+shadow_i,17:21,0], self.bodypoints[fr+shadow_i,17:21,1], 
                             c=self.bp_palette[5], alpha=alpha,
                             marker="o", markersize=1)
                    self.ax1.plot(self.bodypoints[fr+shadow_i,21:24,0], self.bodypoints[fr+shadow_i,21:24,1], 
                             c=self.bp_palette[6], alpha=alpha,
                             marker="o", markersize=1)
                    self.ax1.plot(self.bodypoints[fr+shadow_i,24:27,0], self.bodypoints[fr+shadow_i,24:27,1], 
                             c=self.bp_palette[7], alpha=alpha,
                             marker="o", markersize=1)
                    self.ax1.plot(self.bodypoints[fr+shadow_i,27:30,0], self.bodypoints[fr+shadow_i,27:30,1], 
                             c=self.bp_palette[8], alpha=alpha,
                             marker="o", markersize=1)
                    
                
                # angle plot
                self.ax2.clear()
                self.ax2.set_xlim([start,stop-1]); self.ax2.set_ylim([-1.0,1.0]);
                for i in range(num_angles):
                    self.ax2.plot(np.arange(start, fr+1), self.angles[start:fr+1,i], alpha=0.5, 
                                  linewidth=0.5, label=f"ang {i}", c=self.bp_palette[i+1])
                self.ax2.legend()
                self.ax2.set(xlabel='Frame', ylabel='Normalized Angle')
                
                # ethogram plot
                self.ax3.clear()
                self.ax3.set_xlim([start,stop-1]); self.ax3.set_ylim([-1.5,num_clusters]);
                self.ax3.scatter(np.arange(start, fr+1), self.cluster[start:fr+1], c=self.cluster_colors[start:fr+1], 
                            alpha=1, s=2, marker="s")
                self.ax3.set_yticks(range(-1,num_clusters))
                self.ax3.set(xlabel='Frame', ylabel='Ethogram')
                
                # power spectrogram plot
                self.ax4.clear()
                for i in range(num_angles):
                    self.ax4.plot(self.freq, self.power[i,:,fr].T, label=f"ang {i}", 
                                  alpha=0.5, linewidth=1, c=self.bp_palette[i+1])
                self.ax4.set_xlabel("freq"); self.ax4.set_ylabel("power")
                self.ax4.set_ylim([-0.05,0.8])
                self.ax4.legend()
                
                # scatter cluster plot
                scatter_embed = self.embed[start:fr+1,:]
                scatter_good_fr = np.unique(np.where(~np.isnan(scatter_embed))[0])
                
                self.ax5.clear()
                self.ax5.scatter(self.embed[idx,0], self.embed[idx,1], 
                        c=self.cluster_colors[idx], 
                        alpha=0.2, s=0.05)
                self.ax5.plot(scatter_embed[scatter_good_fr,0], scatter_embed[scatter_good_fr,1],
                        c='k', linewidth=1, alpha=0.5)
                if self.embed[fr,0] is not np.nan:
                    self.ax5.scatter(self.embed[fr,0], self.embed[fr,1],
                            c='k', s=5, marker="x")
                self.ax5.set(xlabel='Component 1', ylabel='Component 2', title=f"frame {fr}", xlim=self.ax6.get_xlim(), ylim=self.ax6.get_ylim())
                
                # take snapshot
                writer.grab_frame()
            writer.grab_frame()
            plt.close()
```

```python
num_cluster = np.max(gmm_label)+1
num_files = len(files_ref)

proportion_matrix = np.zeros((len(files_ref), num_cluster))
row_name = []

for files_idx, (path, fr_range) in enumerate(files_ref.items()):
    # Parse Files
    base = os.path.basename(path)
    basename = base.split("DeepCut")[0]
    row_name.append(basename)
    cluster = gmm_label[fr_range[0]:fr_range[1]]
    # Create Proportion Matrix
    for clust_i in range(num_cluster):
        proportion_matrix[files_idx, clust_i] = len(np.where(cluster == clust_i)[0])/len(cluster)

# Proportion Figure
fig, ax = plt.subplots(figsize=(20,5)) 
ax.imshow(proportion_matrix.T, cmap="jet")

row_name_mod = [row.replace("_", "\n").replace("-", "\n").replace("2018", "18") for row in row_name]

plt.xticks(np.arange(num_files), row_name_mod, rotation=0)
ax.set(ylabel='Clusters', xlabel='Files', 
                title=f"Proportion Matrix")
plt.tight_layout(pad=0)
plt.show()
```
