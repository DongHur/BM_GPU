```python
import os
import pandas as pd
import numpy as np
```

```python
CLUSTER_PATH = "/Users/lukefoster/Documents/CompEtho/BM_GPU/results/cluster"
```

```python
all_clusters = [CLUSTER_PATH + "/" + f for f in os.listdir(CLUSTER_PATH)]
```

```python
cluster_data = {}

for file in all_clusters:
    cluster_data[f"{file.split('cluster_')[-1].split('.npy')[0]}"] = np.load(file)
```

```python
cluster_data
```

```python
def find_max(data):
    max_values = []

    for key, value in cluster_data.items():
        max_values.append(max(value))

    return int(max(max_values))
```

```python
def find_min(data):
    min_values = []

    for key, value in cluster_data.items():
        min_values.append(min(value))

    return int(min(min_values))
```

```python
cluster_values = np.arange(find_min(cluster_data), find_max(cluster_data)+1)
```

```python
cluster_values
```

```python
arr = np.empty((len(cluster_data), len(cluster_values)+1), dtype="object")

for i, dict in enumerate(cluster_data.items()):
    # key = dict[0]
    # value = dict[1]
    row = np.empty(len(cluster_values)+1, dtype="object")
    
    row[0] = dict[0]
    
    # row[1] = count of cluster -1
    # row[2] = count of cluster 0
    # row[3] = count of cluster 1
    # row[4] = count of cluster 2
    # row[n] = count of cluster n-2 ...
    # ... row[67] = count of cluster 65
    
    values = dict[1].tolist()
    
    for value in values:
        value = int(value)
    
    # number of frames for video = len(dict[1])
    
    for j in range(1,68):
        row[j] = values.count(j-2) / len(dict[1])

    arr[i] = row
```

```python
# convert array into dataframe
DF = pd.DataFrame(arr)

DF = DF.rename(columns={0: "Video File"})

for i in range(1,68):
    DF = DF.rename(columns={i: f"Cluster {i-2}"})
```
