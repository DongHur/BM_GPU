```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
config = {
    "skeleton": [
    [0,1],[0,2],[1,2],[6,7],
    [6,8],[8,12],[12,10],[10,11],[11,7]],
    "skeleton_color": [
    'tab:blue','tab:blue','tab:blue','tab:blue',
    'tab:blue','tab:blue','tab:blue','tab:blue','tab:blue']
}
```

```python
data = np.load("keypoint.npy")
```

```python
num_fr, num_dim = data.shape
```

```python
num_bp = 15
data_reformat = data.reshape(num_fr, num_bp, int(num_dim/num_bp))
```

```python
fr = 1000

fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(data_reformat[fr,:,0], data_reformat[fr,:,1])

for side in config['skeleton']:
    ax.plot(data_reformat[fr,side,0], data_reformat[fr,side,1], c='black')
plt.show()
```
